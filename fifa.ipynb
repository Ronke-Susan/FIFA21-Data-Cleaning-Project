### Data Cleaning Processes

#### Removing unwanted characters and creating new columns

```python
#removing newline characters from team&contract and hits columns
fifa['Team & Contract']=fifa['Team & Contract'].str.strip('\n').str.split('\n')
fifa['Hits']=fifa['Hits'].str.strip('\n')
```

<font color='blue'>the split method separated team and contract into separate elements in a list. i then used list indexing to extract them into 
separate columns named team and contract_period respectively</font>

```python
#creating new columns from Team&Contract column
fifa['Team'] =fifa['Team & Contract'].apply(lambda x:x[0])

fifa['Contract_period'] = fifa['Team & Contract'].apply(lambda x:x[1]).str.replace("~","-")
```

<font color='blue'>Next, I repositioned the team and contract_period columns next to the team&contract column before dropping the team&contract column so that instead of having the newly created columns at the end of the dataframe, they can be located where the parent(team&contract) column used to be.</font>

```python
#positioning new team and contract_period columns next to Team&Contract column
#identifying the index of 'Team & Contract' column
t_index= fifa.columns.get_loc('Team & Contract')

#extracting new columns from their positions at the end of the dataframe
new_columns= fifa.iloc[:,-2:]

#dropping new columns from their initial positions at the end of the dataframe
fifa= fifa.iloc[:,:-2]

#postioning the new_columns next to the parent column
fifa= pd.concat([fifa.iloc[:,:t_index+1],new_columns,fifa.iloc[:,t_index+1:]],axis=1)
```

<font color='blue'>Dropped the photoUrl and playerUrl columns since the links were longer available</font>

```python
#dropped photoUrl,playerUrl,Team&Contract columns
fifa.drop(columns=['photoUrl','playerUrl','Team & Contract'],inplace= True)
```

```python
#renaming columns
fifa.rename(columns={'↓OVA':'OVA','Weight':"Weight(lbs)","Value":"Value(€)","Wage":"Wage(€)","Release Clause":"Release Clause(€)"},inplace=True)
```

```python
#removed unwanted characters from the data
replace_dict= {"€":"","★":"","lbs":""}
fifa= fifa.replace(replace_dict, regex=True )
```

<font color='blue'>converted the shorthand notation for values in the value,wage and release clause columns into full numerical values by multiplying each value with its respective million/thousand using a dictionary_based replacement that replaces 'M' with 'e6' and 'K' with 'e3' and converted them to floats using astype function</font> 

```python
#perfomed necessary aggregations on the monetary columns
fifa[['Value(€)','Wage(€)','Release Clause(€)']]=fifa[['Value(€)','Wage(€)','Release Clause(€)']].replace({'M':'e6', 'K':'e3'},regex=True).astype(float)
```

```python
fifa['Hits'].unique()
```

```python
#creating a function to convert values from '1.2K' format in the hits column to thousands
def convert_hits(value):
    if value[-1].lower() == 'k':
        return int(float(value[:-1]) * 1000)
    else:
        return int(value)
    
#applying the function to the hit
fifa['Hits']= fifa['Hits'].apply(convert_hits)
```

```python
fifa['Hits'].unique()
```

#### Handling inconsistencies in the contract period column

<font color='blue'> Not all values in the contract_period column are in the same format, for instance there are values like "2021-2022", "Free", "jan 30,2021 on loan" and I would like to maintain the format "2021-2022" that accurately describes a time period</font>

```python
fifa['Contract_period'].unique()
```

```python
#Removing letters and numbers from before a comma(including the comma),"on loan", and "Free"
fifa['Contract_period']=fifa['Contract_period'].str.replace(r'^[^,]*,|\bOn Loan\b|\bFree\b',"",regex=True).str.rstrip()
```

```python
fifa['Contract_period'].unique()
```

<font color='blue'>There are some missing start/joined date or maybe end date values in the contract_period column, by comparing the joined, loan date end and contract_period columns, I can confirm what date value is missing. Since the loan end date column contain a large number of missing values, I created a conditon that can help filter out the missing values.</font>.  

```python
#comparing contract period with joined date and loan end date
#creating a condition that can help filter the dataframe
filt_er= fifa['Loan Date End'].notna()

#displaying filtered columns
fifa.loc[filt_er,['Joined','Loan Date End','Contract_period']]
```

<font color='blue'>From the filtered dataset, it is obvious that the values missing from the contract_period column are joined year values.Next, I created another filter condition that locates rows where contract_period has only end year values, I then extracted year values from the joined column and added them to their respective contract_period values</font> 

```python
#creating a filter condition that locates where contract period has only an end year
condition= fifa['Contract_period'].str.len() <9
```

```python
#extracted year from Joined date and added to rows where joined year was missing from the contract period
fifa.loc[condition,'Contract_period']=fifa.loc[condition, 'Joined'].str[-4:]+' -'+fifa.loc[condition,'Contract_period']
```

```python
#confirming that missing joined year has been filled
fifa.loc[filt_er,['Joined','Loan Date End','Contract_period']]
```

<font color='blue'>Further inspection of the contract_period column revealed that some rows have missing end dates and these missing end dates are also not present in the loan end date column as shown below.In order to avoid making assumptions by assigning a default value like 'present' indicating that the players' contract is still ongoing, I choose to leave the end dates empty. This is to help preserve the data's integrity and avoid introducing potentially inaccurate and misleading information</font>

```python
#further inspection of the contract_period column
condition= fifa['Contract_period'].str.len() <9
filt_er= fifa['Loan Date End'].notna()
fifa.loc[condition,['Joined','Loan Date End','Contract_period']]
```

```python
#drop Loan Date End column
fifa.drop(columns= 'Loan Date End',inplace=True)
```

#### Creating new columns and changing columns data type

```python
#creating a new height column showing athlete heights in centimeters by converting the height column from feet&inches
def convert_height(height):
    feet,inches = height.split("'")
    total_inches = int(feet) * 12 + int(inches[:-1])
    cm= round(total_inches * 2.54,1)
    return cm

#applying the conversion function to the height column and creating a new column
fifa['Height(cm)']= fifa['Height'].apply(convert_height)
```

```python
#repositioning new height column
#identifying the index of 'Height' column
index= fifa.columns.get_loc('Height')

new_height= fifa.iloc[:,-1:]
#drop the new height position created at the end of the data frame
fifa= fifa.iloc[:,:-1]

#position new height_cm next to height
fifa= pd.concat([fifa.iloc[:,:index+1],new_height,fifa.iloc[:,index+1:]],axis=1)
```

```python
#Creating new columns and changing data types of some columns
fifa['Joined']=pd.to_datetime(fifa['Joined'])
fifa['Month'] = fifa['Joined'].dt.month_name()
fifa['Day_Joined'] = fifa['Joined'].dt.day_name()
fifa[['W/F','SM','IR','Weight(lbs)']]=fifa[['W/F','SM','IR','Weight(lbs)']].astype(np.int64)
```

```python
#repositioning columns created from joined column
j_index= fifa.columns.get_loc('Joined')
new_cols= fifa.iloc[:,-2:]

fifa= fifa.iloc[:,:-2]
fifa= pd.concat([fifa.iloc[:,:j_index+1],new_cols,fifa.iloc[:,j_index+1:]],axis=1)
```

#### Handling Duplicates

<font color= 'blue'>In this dataset, each player has a unique ID. To find duplicates in this dataset, I used the ID column as the subset parameter for the *duplicated()* function to filter and display the duplicate rows. next, I proceeded to drop one of the duplicate rows by using *drop_duplicates()* function and setting the keep parameter to first(this retains the first occurence of the duplicate rows and removes subsequent one)</font>

```python
#finding duplicates in the data
duplicates = fifa[fifa.duplicated(subset='ID',keep=False)]
duplicates
```

```python
#dropping duplicates in the data
fifa.drop_duplicates(subset='ID', keep ='first',inplace= True)
```
